1.What is a Vector Database (VectorDB) and how is it different from traditional databases?

Ans: A Vector Database stores high-dimensional embeddings and performs similarity search using cosine similarity or Euclidean distance, unlike traditional SQL databases which perform exact-match queries on structured data.


2.Explain the various types of VectorDBs available and describe their suitability for different use cases.

Ans: Types of Vector Databases:
- Dedicated Vector DBs (Pinecone, Milvus, Chroma) used for Semantic search
- Hybrid DBs (PostgreSQL + pgvector) used for RAG
- In-memory stores (FAISS) used for Recommendation systems

```python
#6 Install and set up Chroma DB, and insert sample vector data for semantic search.
# Install required libraries
!pip install chromadb sentence-transformers

import chromadb
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')
client = chromadb.Client()
collection = client.create_collection('demo')

docs = ['AI is intelligence by machines',
        'Machine learning is a subset of AI',
        'Deep learning uses neural networks']

embeddings = model.encode(docs).tolist()

collection.add(documents=docs, embeddings=embeddings, ids=['1','2','3'])

results = collection.query(query_texts=['What is AI?'], n_results=2)
print(results)
```

```python
#  Demonstrate how to download and fine-tune a Hugging Face model for a text generation task.
!pip install transformers datasets

from transformers import GPT2Tokenizer, GPT2LMHeadModel

model_name = 'gpt2'
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

input_ids = tokenizer('AI is transforming the world', return_tensors='pt')
output = model.generate(**input_ids, max_length=50)
print(tokenizer.decode(output[0]))
```

```python
#9 Implement a basic RAG (Retrieval-Augmented Generation) system using Ollama with Llama3.


query = 'Explain AI'
docs = collection.query(query_texts=[query], n_results=2)
context = ' '.join(docs['documents'][0])

response = ollama.chat(
    model='llama3',
    messages=[
        {'role': 'system', 'content': 'Use context to answer.'},
        {'role': 'user', 'content': f'Context: {context}\nQuestion: {query}'}
    ]
)

print(response['message']['content'])
```
